```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df  = pd.read_csv('../pilrek.csv', delimiter = ',', names = ['Timestamp', 'Kategori', 'Umur', 'CaRek', 'Alasan', 'Harapan'])
df
```

```python
sort_alasan = df[df.Alasan == 'Sesuai dengan kriteria kekinian (Rektor 4.0)']
groupby_carek = sort_alasan.groupby('CaRek')
count_sorted = groupby_carek.count().sort_values('Timestamp', ascending = False)
carek_kekinian = count_sorted[['Timestamp']]
```

```python
plot_2 = carek_kekinian.plot(kind='bar')
plot_2.set_xlabel('')
plot_2.set_title('Carek memiliki kriteria kekinian (Rektor 4.0)')
```
