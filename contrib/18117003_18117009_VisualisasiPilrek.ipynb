```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url1 = 'https://github.com/eueung/pilrek/blob/master/pilrek-anon.csv'
url2 = 'https://github.com/eueung/pilrek/blob/master/pilrek.csv'
#url1 = 'pilrek.csv'
#url2 = 'pilrek-anon.csv'
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df = pd.concat([df1,df2])
```

```python
#1 Visualisasi Alasan Pemilih Carek STEI
df_stei = df[df['CaRek Pilihan'].str.contains('STEI')]
df_stei_baru = df_stei.dropna()
```

```python
plotSTEI = df_stei_baru['Alasan Memilih CaRek'].value_counts().plot(kind = 'bar', color = ['black','pink']) 
plotSTEI.set_ylabel('Frekuensi')
plotSTEI.set_title('Alasan Memilih CaRek STEI')
```

```python
#2 Visualisasi harapan alumni memilih CaRek
df_alumni = df[df['Kategori Anda'].str.contains('Alumni')]
df_alumni_bersih = df_alumni.dropna()
df_alumni_bersih
```

```python
plotAlumni = df_alumni_bersih['Alasan Memilih CaRek'].value_counts().plot(kind = 'bar',color =['cyan','orange','red','grey','magenta','black','blue']) 
plotAlumni.set_ylabel('Frekuensi')
plotAlumni.set_title('Alasan Alumni Memilih CaRek')
```

```python
#3 Visualisasi CaRek yang dipilih karena Keberhasilan dan prestasi
df_carektor = df[df['Alasan Memilih CaRek'].str.contains('Keberhasilan dan prestasi')]
df_carektor
plotCaRek = df_carektor['CaRek Pilihan'].value_counts().plot(kind = 'bar', color =['green','yellow','red','aqua','magenta','pink','brown'] )
plotCaRek.set_ylabel('Frekuensi')
plotCaRek.set_title('CaRek yang dipilih karena Keberhasilan dan prestasi')
```
