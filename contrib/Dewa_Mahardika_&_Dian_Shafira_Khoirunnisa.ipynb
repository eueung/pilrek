```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
#Impor Dataset
file = '../pilrek.csv'
data_pilrek = pd.read_csv(file)
data_pilrek ##Cek Datanya
```

```python
##SAJIAN DATA 1 : Double Donut Jumlah Kategori Pemilih dan Umurnya
```

```python
#Parsing Data Kategori ke List
data_kategori = data_pilrek['Kategori Anda'].value_counts()
data_kategori
```

```python
group_names = data_kategori.keys().tolist()
group_names
```

```python
group_sizes = data_kategori.tolist()
group_sizes
```

```python
#Parsing Data Umur tiap Kategori ke List
```

```python

data_umur_alumni  = data_pilrek[data_pilrek['Kategori Anda']=='Alumni']['Kategori Umur Anda'].value_counts()
data_umur_alumni
```

```python
data_umur_umum  = data_pilrek[data_pilrek['Kategori Anda']=='Umum']['Kategori Umur Anda'].value_counts()
data_umur_umum
```

```python
data_umur_mahasiswa  = data_pilrek[data_pilrek['Kategori Anda']=='Mahasiswa']['Kategori Umur Anda'].value_counts()
data_umur_mahasiswa
```

```python
data_umur_dosen  = data_pilrek[data_pilrek['Kategori Anda']=='Dosen']['Kategori Umur Anda'].value_counts()
data_umur_dosen
```

```python
data_umur_pegawai  = data_pilrek[data_pilrek['Kategori Anda']=='Pegawai / Tendik']['Kategori Umur Anda'].value_counts()
data_umur_pegawai
```

```python
subgroup_names = data_umur_alumni.keys().tolist() + data_umur_umum.keys().tolist() +data_umur_mahasiswa.keys().tolist() + data_umur_dosen.keys().tolist() + data_umur_pegawai.keys().tolist() 
subgroup_names
```

```python
subgroup_sizes = data_umur_alumni.tolist() + data_umur_umum.tolist() +data_umur_mahasiswa.tolist() + data_umur_dosen.tolist() + data_umur_pegawai.tolist() 
subgroup_sizes
```

```python
#Verifikasi Data
tabel_kategori_dan_umur = data_pilrek.groupby(['Kategori Anda','Kategori Umur Anda']).size().unstack(fill_value=0).reset_index()
tabel_kategori_dan_umur
```

```python
#Warna
a,b,c,d,e =  [plt.cm.Blues, 
              plt.cm.Reds, 
              plt.cm.Greens, 
              plt.cm.Oranges, 
              plt.cm.Purples]
```

```python
#Lingkaran Luar
fig,ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_sizes, radius=3, 
                  labels=group_names, 
                  colors=[a(0.45), 
                          b(0.45), 
                          c(0.45),
                          d(0.45),
                          e(0.45)]
                  )
plt.setp( mypie, width=0.7, edgecolor='pink')
#Lingkaran Dalam
mypie2, _ = ax.pie(subgroup_sizes, 
                   radius=3-0.7, 
                   labels=subgroup_names, 
                   labeldistance=0.8, 
                   colors=[a(0.3), 
                           a(0.1), 
                           a(0.2),
                           b(0.3), 
                           b(0.2),
                           b(0.2),
                           c(0.3), 
                           c(0.3), 
                           c(0.1),
                           d(0.3), 
                           d(0.3), 
                           d(0.1),
                           e(0.3), 
                           e(0.3), 
                           e(0.1)]
                  )
plt.setp( mypie2, width=0.7, edgecolor='white')
plt.margins(0,0)
plt.show()
```
