```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/eueung/pilrek/raw/master/pilrek-anon.csv'
df  = pd.read_csv(url, delimiter=',', names = ['Timestamp', 'Kategori_Anda', 'Kategori_Umur_Anda', 'CaRek_Pilihan', 'Alasan_Memilih_CaRek', 'Harapan_Anda'])
```

```python
#VISUALISASI 1
#Mencari Calon Rektor Pilihan untuk kategori Dosen dan Mahasiswa (Internal ITB) dari segala umur dan divisualisasikan melalui bar chart

#Mem-filter responden untuk kategori Dosen dan Mahasiswa yang digabungkan ke dalam variable Internal
dosen=df[df.Kategori_Anda == 'Dosen']
mahasiswa=df[df.Kategori_Anda == 'Mahasiswa']

internal=pd.concat([dosen,mahasiswa])
internal
```

```python
#Jumlah dosen yang memilih setiap Calon Rektor dalam variabel d

d =dosen.groupby(['CaRek_Pilihan']).count()
d
```

```python
#Mengurutkan 5 Calon Rektor Pilihan berdasarkan jumlah pemilih terbanyak dari kategori dosen

d[['Kategori_Anda']].sort_values(by = ['Kategori_Anda']).tail(5)
```

```python
#Jumlah mahasiswa yang memilih setiap Calon Rektor dalam variabel m

m =mahasiswa.groupby(['CaRek_Pilihan']).count()
m
```

```python
#Mengurutkan 5 Calon Rektor Pilihan berdasarkan jumlah pemilih terbanyak dari kategori mahasiswa

m[['Kategori_Anda']].sort_values(by = ['Kategori_Anda']).tail(5)
```

```python
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
#Membuat visualisasi data di atas melalui multiple bar-chart

n_groups = 7
Internal_dosen = (0, 17,55,35, 0, 22,22)
Internal_mahasiswa = (54, 0, 46, 57, 25, 0, 20)
```

```python
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.3
opacity = 1

rects1 = plt.bar(index, Internal_dosen, bar_width, alpha=opacity, color='b',label='Dosen')
rects2 = plt.bar(index + bar_width, Internal_mahasiswa, bar_width, alpha=opacity, color='r', label='Mahasiswa')

plt.xlabel('Calon Rektor Pilihan')
plt.ylabel('Jumlah Pemilih')
plt.title('Internal ITB')
plt.xticks(index + bar_width/2, ('Acep Purqon (FMIPA-FI)', 'Dicky Rezady Munaf (FSRD)', 'Edy Tri Baskoro (FMIPA-MA)', 'Kadarsah Suryadi (FTI-TI)', 'Khairurrijal (FMIPA-FI)', 'Sigit Puji Santosa (FTMD)', 'Tutuka Ariaji (FTTM-TM)'), rotation=90, fontsize=7)
plt.legend()
           
    
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
#VISUALISASI 2

#Mencari Calon Rektor Pilihan untuk kategori Umum, Alumni, dan Pegawai (Eksternal ITB) dari segala umur dan divisualisasikan melalui bar chart

#Mem-filter responden untuk kategori alumni, pegawai, dan umum yang digabungkan ke dalam variable eksternal

alumni=df[df.Kategori_Anda == 'Alumni']
pegawai=df[df.Kategori_Anda == 'Pegawai / Tendik']
umum=df[df.Kategori_Anda == 'Umum']

eksternal=pd.concat([alumni,pegawai,umum])
eksternal
```

```python
#Jumlah pegawai yang memilih setiap Calon Rektor dalam variabel p

p =pegawai.groupby(['CaRek_Pilihan']).count()
p
```

```python
#Mengurutkan 5 Calon Rektor Pilihan berdasarkan jumlah pemilih terbanyak dari kategori pegawai

p[['Kategori_Anda']].sort_values(by = ['Kategori_Anda']).tail(5)
```

```python
#Jumlah masyarakat umum yang memilih setiap Calon Rektor dalam variabel u

u =umum.groupby(['CaRek_Pilihan']).count()
u
```

```python
#Mengurutkan 5 Calon Rektor Pilihan berdasarkan jumlah pemilih terbanyak dari kategori masyarakat umum

u[['Kategori_Anda']].sort_values(by = ['Kategori_Anda']).tail(5)
```

```python
#Membuat visualisasi data di atas melalui multiple bar-chart

n_groups = 9
Eksternal_alumni = (81, 0, 71, 54, 0, 83, 0, 0, 205)
Eksternal_pegawai = (0, 20, 0, 17, 0, 82, 0, 13, 24)
Eksternal_umum = (78, 68, 0, 0, 56, 67, 48, 0, 0)

fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.3
opacity = 1

rects1 = plt.bar(index-bar_width, Eksternal_alumni, bar_width, alpha=opacity, color='r',label='Alumni')
rects2 = plt.bar(index, Eksternal_pegawai, bar_width, alpha=opacity, color='b', label='Pegawai')
rects3 = plt.bar(index + bar_width, Eksternal_umum, bar_width, alpha=opacity, color='g', label='Umum')

plt.xlabel('Calon Rektor Pilihan')
plt.ylabel('Jumlah Pemilih')
plt.title('Eksternal ITB')
plt.xticks(index + bar_width/10, ('Acep Purqon (FMIPA-FI)', 'Dicky Rezady Munaf (FSRD)', 'Deddy Kurniadi (FTI-TF)', 'Edy Tri Baskoro (FMIPA-MA)', 'Emir Mauludi Husni (STEI)', 'Kadarsah Suryadi (FTI-TI)', 'Khairurrijal (FMIPA-FI)', 'Sigit Puji Santosa (FTMD)', 'Tutuka Ariaji (FTTM-TM)'), rotation=90, fontsize=7)
plt.legend()
           
    
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 3, height),
                    xytext=(0, 4),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()

#KESIMPULAN
dari Internal ITB, Calon Rektor Pilihan yang unggul adalah Pak Edy dengan 101 pemilih, sedangkan di kalangan Eksternal ITB, Pak Kadarsah lebih unggul dengan 232 pemilih
```
